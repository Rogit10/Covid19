```python
# Dropping columns that are not needed

covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace = True, axis = 1)
```

```python
covid_df.head(10)
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format='%m/%d/%Y')
```

```python
covid_df.head()
```

```python
# Active Cases
covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured']
                                                    + covid_df['Deaths'])
covid_df.tail()
```

```python
# Creating a pivot table with selected columns

statewise = pd.pivot_table(covid_df, values = ['Confirmed', 'Deaths', 'Cured'],
                           index = 'State/UnionTerritory', aggfunc= max)

# Calculating the recovery rate
statewise ['Recovery_Rate'] = statewise['Cured']*100/statewise['Confirmed']

# Calculating the mortality rate
statewise ['Mortality_Rate'] = statewise['Deaths']*100/statewise['Confirmed']

# Sorting the Data by 'Confirmed cases' and Mapping it
statewise = statewise.sort_values(by = 'Confirmed', ascending = False )
statewise.style.background_gradient(cmap = 'viridis')
```

```python
# Top 10 States with Active cases
Top_10_active = covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases', 'Date']].sort_values(by=['Active_Cases'], ascending=False).reset_index()
fig = plt.figure(figsize = (16,9))

# Using seaborn to plot a bar chart
ax = sns.barplot(Top_10_active.iloc[:10], y='Active_Cases', x='State/UnionTerritory', linewidth=2, edgecolor='blue')
plt.xlabel("States")
plt.ylabel("Active Cases")
plt.title("Top 10 states with Active Cases",size = 20)
plt.show()
```

```python
#Top 10 States with Death rate

Top_10_death = covid_df.groupby(by='State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by=['Deaths'], ascending=False).reset_index()
fig = plt.figure(figsize = (16,9))

# Using seaborn to plot a bar chart
ax = sns.barplot(Top_10_death.iloc[:12], y='Deaths', x='State/UnionTerritory', linewidth=2, edgecolor='blue')
plt.xlabel("States")
plt.ylabel("Deaths")
plt.title("Top 10 states with Death Cases",size = 20)
plt.show()
```

```python
# Growth Trend for the Top 5 affected states
fig = plt.figure(figsize=(20, 9))

ax = sns.lineplot(
    data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh'])],
    x='Date',
    y='Active_Cases',
    hue='State/UnionTerritory'
)
ax.set_title("Top 5 Affected States in India", size=16)
ax.set_xlabel('Date')
```

```python
# Removing columns that have too many null values
vaccination = vaccine_df.drop( columns = ['Sputnik V (Doses Administered)', 'AEFI','18-44 Years (Doses Administered)',
                                          '45-60 Years (Doses Administered)', '60+ Years (Doses Administered)'], axis = 1)
vaccination.head()
```

```python
# Male vs Female

male = vaccination['Male (Doses Administered)'].sum()
female = vaccination['Female (Doses Administered)'].sum()

px.pie(names = ["Male","Female"], values = [male,female], title = "Male and Female Vaccination")
```

```python
# Removing 'india' from the state column

vaccine = vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
# Most vaccinated State

max_vac = vaccine.groupby('State')['Total Individuals Vaccinated'].sum().to_frame('Total Individuals Vaccinated')
max_vac = max_vac.sort_values('Total Individuals Vaccinated', ascending = False)[:5]
```

```python
# Bar chart for top 5 vaccinated states
fig = plt.figure(figsize = (10,5))
plt.title("Top 5 vaccinated States in India", size = 20)
x = sns.barplot(max_vac.iloc[:10], y= 'Total Individuals Vaccinated', x= max_vac.index, linewidth=2, edgecolor='blue')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
